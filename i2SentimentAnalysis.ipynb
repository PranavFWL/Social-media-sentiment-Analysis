```python
rw = ['nan', 'virginamerica', 'americanair', 'usairway', 'jetblu', 'southwestair', 'unit']

from tkinter.constants import Y
def tars_text(text):
  text = text.lower()
  text = nltk.word_tokenize(text)

  y = []  #Removing special chars
  for i in text:
    if i.isalnum():
      y.append(i)

  text = y[:]
  y.clear()

  for i in text: #Removing stop and punctuations
    if i not in sw.words('english') and i not in sr.punctuation and i not in rw:
      y.append(i)

  text = y[:]
  y.clear()

  for i in text:
    y.append(PorterStemmer().stem(i))
  return " ".join(y)
```

```python
dsn['text'] = dsn['text'].apply(lambda x: tars_text(x))
```

Vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
```

model training

```python
from sklearn.model_selection import train_test_split
from tensorflow.keras import Sequential
from keras.callbacks import ModelCheckpoint
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.layers import Dense, Dropout, LSTM
from tensorflow import keras
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
from sklearn.ensemble import RandomForestClassifier
model = Sequential()
```

```python
xtrain ,xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.8, random_state = 100)
```

```python
guss = GaussianNB()
multi = MultinomialNB()
berno = BernoulliNB()
rf = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
guss.fit(xtrain, ytrain)
multi.fit(xtrain, ytrain)
berno.fit(xtrain, ytrain)
rf.fit(xtrain, ytrain)
```

```python
predict1 = guss.predict(xtest)
predict2 = multi.predict(xtest)
predict3 = berno.predict(xtest)
predict4 = rf.predict(xtest)
```

```python
print(accuracy_score(predict1, ytest))
print(accuracy_score(predict2, ytest))
print(accuracy_score(predict3, ytest))
print(accuracy_score(predict4, ytest))
print()
print(confusion_matrix(predict1, ytest))
print(confusion_matrix(predict2, ytest))
print(confusion_matrix(predict3, ytest))
print(confusion_matrix(predict4, ytest))
```
